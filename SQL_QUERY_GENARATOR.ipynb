```python
class SQLQueryGenerator:

    def select_query(self, table, columns):
        column_string = ", ".join(columns)
        return f"SELECT {column_string} FROM {table};"

    def insert_query(self, table, data):
        columns = ", ".join(data.keys())
        values = "', '".join(data.values())
        return f"INSERT INTO {table} ({columns}) VALUES ('{values}');"

    def update_query(self, table, data, condition):
        set_values = ", ".join([f"{k}='{v}'" for k, v in data.items()])
        return f"UPDATE {table} SET {set_values} WHERE {condition};"

    def delete_query(self, table, condition):
        return f"DELETE FROM {table} WHERE {condition};"


# Main Program
sql = SQLQueryGenerator()

print(sql.select_query("students", ["id", "name", "age"]))

data = {"id": "1", "name": "Maria", "age": "20"}
print(sql.insert_query("students", data))

update_data = {"age": "21"}
print(sql.update_query("students", update_data, "id=1"))

print(sql.delete_query("students", "id=1"))
```
